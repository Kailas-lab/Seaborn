```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the example CSV file
data = pd.read_csv("data2sea.csv")

# Display the first few rows
print(data.head())

# 1. Scatter Plot
sns.scatterplot(x="x_values", y="y_values", hue="category", data=data)
plt.title("Scatter Plot: x_values vs y_values")
plt.show()







```
