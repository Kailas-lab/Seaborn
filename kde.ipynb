```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the example CSV file
data = pd.read_csv("data2sea.csv")

# Display the first few rows
print(data.head())

sns.kdeplot(data=data, x="value", hue="category", fill=True)
plt.title("KDE Plot: Distribution of 'value' by Category")
plt.show()
```
