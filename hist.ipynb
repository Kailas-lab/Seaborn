```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the example CSV file
data = pd.read_csv("data2sea.csv")

# Display the first few rows
print(data.head())

# 4. Histogram
sns.histplot(data=data, x="value", kde=True, bins=10, hue="category")
plt.title("Histogram: Distribution of Value")
plt.show()
```
