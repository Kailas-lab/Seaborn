```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the example CSV file
data = pd.read_csv("data2sea.csv")

# Display the first few rows
print(data.head())

# 1. regplot (Scatter plot with regression line)
#plt.figure(figsize=(6, 4))
sns.regplot(x="x_values", y="y_values", data=data)
plt.title("Regplot: x_values vs y_values with Regression Line")
plt.show()
```
