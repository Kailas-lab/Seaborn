```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the example CSV file
data = pd.read_csv("data2sea.csv")

# Display the first few rows
print(data.head())

# 2. Line Plot
sns.lineplot(x="x_values", y="value", hue="category", data=data)
plt.title("Line Plot: x_values vs value")
plt.show()
```
