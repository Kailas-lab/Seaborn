## Boxplot

```python
sns.boxplot(x="species",y="sepal_width",data=iris,hue="species")
```

Components of the Box Plot:

Each box plot has the following parts:

1. Median (Central Line inside the box):
This line inside the box represents the median (50th percentile) of sepal width for that species.

2. Box (Interquartile Range, IQR):
The bottom of the box is the 25th percentile (Q1).

The top of the box is the 75th percentile (Q3).

The box spans the IQR = Q3 - Q1, which contains the middle 50% of the data.

3. Whiskers (Vertical lines above and below the box):
These extend to the smallest and largest values within 1.5  IQR from Q1 and Q3.

They give an idea of how spread out the data is beyond the IQR.

4. Outliers (Dots outside the whiskers):
Points that lie outside 1.5  IQR from the quartiles.

These are plotted as individual circles/dots and represent unusually high or low values.

Interpreting the Plot by Species

Setosa
Median is around 3.4.

Wider IQR and higher whisker range.

Outliers: Both low and high (near 2.3 and 4.4).

Most variable sepal width among the three species.

Versicolor
Median around 2.8.

IQR is tighter, indicating less variation.

Outliers present on both ends, especially low values around 2.0.

Narrowest spread.

Virginica
Median also around 3.0.

IQR and whiskers are slightly wider than versicolor but narrower than setosa.

Outliers present, showing occasional extreme values.
